Among employees who left, get the department wise (use column "sales"), number of employees and rank them.

```python
import pandas as pd
df = pd.read_csv("/content/HR_comma_sep (2).csv")
left_employees = df[df['left']]
department_counts = left_employees.groupby('num_employees').size().reset_index(name='count')
department_counts['rank'] = department_counts['count'].rank(ascending=False)
print(department_counts)
```

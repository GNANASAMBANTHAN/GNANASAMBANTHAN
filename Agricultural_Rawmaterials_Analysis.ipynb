```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Sample data for demonstration (replace with actual dataset)
data = {
    'Crop': ['Rice', 'Wheat', 'Cotton', 'Maize', 'Sugarcane'],
    'Yield (kg/hectare)': [3000, 2500, 2200, 4000, 3500],
    'Cost per kg (INR)': [20, 25, 35, 18, 22],
    'Market Price per kg (INR)': [30, 35, 40, 28, 25],
    'Quality Rating': [8, 7, 6, 9, 8]
}

# Creating a pandas DataFrame
df = pd.DataFrame(data)

# Display the first few rows of the dataframe
print(df.head())
```

```python

# Add a new column for profitability per kg
df['Profit per kg (INR)'] = df['Market Price per kg (INR)'] - df['Cost per kg (INR)']

# Calculate total profit per crop (Profit per kg * Yield per hectare)
df['Total Profit (INR)'] = df['Profit per kg (INR)'] * df['Yield (kg/hectare)']

# Display the updated dataframe
print(df)

# Basic Statistical Analysis
print("\nStatistical Summary:")
print(df.describe())
```

```python
# Plotting Yield vs. Market Price
plt.figure(figsize=(10, 6))
sns.barplot(x='Crop', y='Yield (kg/hectare)', data=df, palette='Set2')
plt.title('Crop Yield Comparison')
plt.ylabel('Yield (kg/hectare)')
plt.show()

# Plotting Total Profit per Crop
plt.figure(figsize=(10, 6))
sns.barplot(x='Crop', y='Total Profit (INR)', data=df, palette='Blues')
plt.title('Total Profit per Crop')
plt.ylabel('Total Profit (INR)')
plt.show()
```
